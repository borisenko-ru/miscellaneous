```python
import pandas as pd
```

```python
class FracPlots:
    '''Provides sample plots for frac data.'''

    def __init__(self, data, threshold=0):
        self.data = data
        self.threshold = threshold
        
    def pressureplot(self):
        df = pd.read_csv(self.data, sep=' ')
        df = df.loc[3:]
        num_columns = list(df.columns)[1:]
        df[df.columns[0]]=df[df.columns[0]].apply(pd.to_datetime, format='%d:%m:%y:%H:%M:%S', errors='coerce')
        df[num_columns]=df[num_columns].apply(pd.to_numeric, errors='coerce')
        return df[self.threshold:].plot(x='JobTime', y='TR_PRESS', figsize=(15, 5))
    
    def prcplot(self):
        df = pd.read_csv(self.data, sep=' ')
        df = df.loc[3:]
        num_columns = list(df.columns)[1:]
        df[df.columns[0]]=df[df.columns[0]].apply(pd.to_datetime, format='%d:%m:%y:%H:%M:%S', errors='coerce')
        df[num_columns]=df[num_columns].apply(pd.to_numeric, errors='coerce')
        ax = df[self.threshold:].plot(x='JobTime', y=['TR_PRESS', 'PROP_CON'])
        return df[self.threshold:].plot(x='JobTime', y='SLUR_RATE', secondary_y=True, ylabel='TR_PRESS \ PROP_CON', ax=ax, figsize=(15, 5))
        
```

```python
FracPlots('Test_2.txt').pressureplot()
```

```python
FracPlots('Test_2.txt').prcplot()
```

```python
FracPlots('Test_2.txt', threshold=8000).pressureplot()
```

```python
FracPlots('Test_2.txt', threshold=8000).prcplot()
```
